```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

#load data and get some basic information
recent_grads = pd.read_csv('recent-grads.csv')
print(recent_grads.head(5))
print(recent_grads.tail(5))
```

```python
#clean NaN values
recent_grads.dropna(inplace = True)
recent_grads.shape[0]
recent_grads.describe()
```

```python
# Four axes, returned as a 2-d array
f, axarr = plt.subplots(2, 2, figsize=(8,8))
f.subplots_adjust(hspace=0, wspace=0)
axarr[0, 0].hist(recent_grads['ShareWomen'])
axarr[0, 0].set_ylabel('ShareWomen')
axarr[0, 0].set_ylim(0, 30)
axarr[0, 0].set_yticklabels([0, 5, 10, 15, 20, 25, 30])
axarr[0, 1].scatter(recent_grads['Unemployment_rate'],recent_grads['ShareWomen'])
axarr[0, 1].set_ylim(0.0, 1.0)
axarr[1, 0].scatter(recent_grads['ShareWomen'], recent_grads['Unemployment_rate'])
axarr[1, 0].set_xlabel('ShareWomen')
axarr[1, 0].set_ylabel('Unemployment_rate')
axarr[1, 0].set_xlim(0, 1.0)
axarr[1, 0].set_ylim(0.0, 0.20)
axarr[1, 0].set_yticklabels([0.00, 0.05, 0.10, 0.15]) 
axarr[1, 0].set_xticklabels([0.0, 0.2, 0.4, 0.6, 0.8], rotation=90) 
axarr[1, 1].hist(recent_grads['Unemployment_rate'])
axarr[1, 1].set_xlabel('Unemployment_rate')
axarr[1, 1].set_xlim(0.0, 0.20)
axarr[1, 1].set_xticklabels([0.00, 0.05, 0.10, 0.15, 0.20], rotation=90)
# Fine-tune figure; hide x ticks for top plots and y ticks for right plots
plt.setp([a.get_xticklabels() for a in axarr[0, :]], visible=False)
plt.setp([a.get_yticklabels() for a in axarr[:, 1]], visible=False)
```

```python
#Bar plots to review gender data with Pandas
arts = recent_grads[recent_grads['Major_category'] == 'Arts']
arts.set_index('Major', inplace=True)
arts[['Men', 'Women']].plot(kind='bar')
```

```python
#setup and build graph in matplotlib
import numpy as np
recent_grads['ShareMen'] = recent_grads['Men']/recent_grads['Total']
arts = recent_grads[recent_grads['Major_category'] == 'Arts']
locs = np.arange(0, len(arts['Major'].unique()), 1)
width = 0.35
offset_locs = locs + width


#plot details
fig_rg2 = plt.figure(figsize=(8,8))
ax_rg2 = fig_rg2.add_subplot(111)


#Build bars and axis details
ax_rg2_bar = ax_rg2.bar(locs, arts['ShareMen'].tolist(), width)
ax_rg2_bar2 = ax_rg2.bar(offset_locs, arts['ShareWomen'].tolist(), width, color='green')
ax_rg2.set_xticklabels(arts['Major'].tolist(), rotation=90)
ax_rg2.set_xticks(offset_locs)

#Display legend, grid and plot
plt.legend((ax_rg2_bar, ax_rg2_bar2), ('ShareMen','ShareWomen'), loc='upper left')
plt.grid()
plt.show()

```
