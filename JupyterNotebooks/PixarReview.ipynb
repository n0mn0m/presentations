```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

#Review the dataframe
pixar_movies = pd.read_csv('PixarMovies.csv')
print(pixar_movies.shape[0])
print(pixar_movies.shape[1])
print(pixar_movies.head(3))
print(pixar_movies.info())

#Some Dataframe Cleanup
pixar_movies['Domestic %'] = pixar_movies['Domestic %'].str.rstrip('%').astype('float')
pixar_movies['International %'] = pixar_movies['International %'].str.rstrip('%').astype('float')
pixar_movies['IMDB Score'] = pixar_movies['IMDB Score'] * 10
filtered_pixar = pixar_movies[pixar_movies.Movie != 'Inside Out']
```

```python
#Plot some rating trends
plt.figure(figsize=(10,6))
plt.plot(filtered_pixar[['RT Score','IMDB Score','Metacritic Score']])
plt.xticks(filtered_pixar.index, filtered_pixar['Movie'], rotation=45)
plt.show()
```

```python
#Look for bias
plt.figure(figsize=(9,5))
critics_reviews = filtered_pixar[['RT Score','IMDB Score','Metacritic Score']]
plt.boxplot(critics_reviews.values)
plt.xticks([1,2,3], critics_reviews.columns.values)
plt.show()
```

```python
#Review domestic and international 
revenue_proportions = filtered_pixar[["Domestic %", "International %"]]
revenue_proportions.plot(kind='bar', stacked=True, figsize=(7,6))
plt.show()
```
